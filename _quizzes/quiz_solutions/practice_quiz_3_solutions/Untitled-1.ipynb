```python
############################################################
# INTRO PYTHON + PANDAS QUIZ SCRIPT 

print("I will abide by Emory's code of conduct")

# import necessary libraries here
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

########################
# 0. RULES

## pandas column operations → ALWAYS row-wise (by index)
## filtering → df[condition]
## groupby → df.groupby("col")["target"].mean()

## use & and | ONLY (never "and"/"or")
## iloc → [row, column] (ROWS FIRST, COLUMNS SECOND)

## = assigns value
## == checks equality


########################
# 1. VARIABLE CREATION
x = 5                     ## create variable
y = x + 2                 ## new variable → y = 7

x = x + 1                 ## reassignment → updates x

flag = x > 5              ## boolean → True / False


########################
# 2. LISTS + APPEND
lst = [1, 2, 3]

lst.append(4)
## adds ONE element to end → [1,2,3,4]
## modifies list IN PLACE
## returns None

lst.append([5,6])
## adds list as ONE element → [1,2,3,4,[5,6]]

lst.extend([5,6])
## adds elements individually → [1,2,3,4,5,6]


########################
# 3. READ CSV
df = pd.read_csv("file.csv")


########################
# 4. INSPECT DATA
df.head()          ## first 5 rows
df.shape           ## (rows, columns)
df.columns         ## column names


########################
# 5. SELECT COLUMNS
df["A"]            ## ONE column (Series)

df[["A","B"]]      ## MULTIPLE columns (DataFrame)

########################
# SUBSETTING

## Select rows by condition
df[df["A"] > 2]

## Select rows AND columns together
df[df["A"] > 2][["A","B"]]
## Step 1: filter rows
## Step 2: keep only selected columns

## Select specific rows by position (iloc)
df.iloc[0]          ## first row
df.iloc[0:3]        ## first 3 rows

## Select rows and columns by position
df.iloc[0:3, 1:3]   ## rows 0–2, columns 1–2

## Select ALL rows, specific column
df.iloc[:, 1]

## Select specific columns by name
df[["A","B"]]

## Key idea:
## filtering = condition-based
## iloc = position-based



########################
# 6. FILTER ROWS

df[df["A"] > 2]
## keep rows where condition is True

df[(df["A"] > 2) & (df["B"] < 6)]
## AND condition

df[(df["A"] > 2) | (df["B"] < 6)]
## OR condition

## wrap each condition in ()
## use & and | (NOT and/or)


########################
# 7. CREATE NEW COLUMN
df["total"] = df["A"] + df["B"]

## row-wise operation:
## total[i] = A[i] + B[i]

df["flag"] = df["A"] > 2
## column of True/False


########################
# 8. GROUPBY
df.groupby("A")["B"].mean()
## mean of B for each value of A

df.groupby("A")["B"].agg(["mean","sum","count"])
## multiple summary statistics


########################
# 9. COUNT VALUES
df["A"].value_counts()
## frequency of each unique value


########################
# 10. SORT
df.sort_values(by="A")
df.sort_values(by="A", ascending=False)


########################
# 11. MISSING DATA


df.isna()
df.isna().sum()

df.dropna()
df.fillna(0)


########################
# 12. APPLY


df["A"].apply(lambda x: x**2)
## applies function to EACH element


########################
# 13. ILOC

df.iloc[rows, columns]

## ALWAYS → [row, column]
## rows FIRST, columns SECOND

df.iloc[0,1]      
## row 0, column 1

df.iloc[0]        
## entire first row

df.iloc[:,1]      
## all rows, column 1

df.iloc[0:3,1:3]  
## rows 0–2, columns 1–2


########################
# 14. COMMON COMBINATIONS


df[df["score"] > 80]["name"]
## filter → then select column

df.groupby("group")["score"].mean()
## group → mean

df["total"] = df["A"] + df["B"]
df[df["total"] > 10]
## create → then filter


########################
# 15. COMMON TRAPS (WITH FIXES)

# WRONG
df["A" > 2]
## comparing string "A" to 2

# CORRECT
df[df["A"] > 2]
## compare values inside column A


# WRONG
df[(df["A"] > 2) and (df["B"] < 5)]
## pandas does NOT use "and"

# CORRECT
df[(df["A"] > 2) & (df["B"] < 5)]
## use & for element-wise AND


# WRONG
df[df["A"] > 2 & df["B"] < 5]
## missing parentheses → order of operations issue

# CORRECT
df[(df["A"] > 2) & (df["B"] < 5)]
## ALWAYS wrap each condition in ( )


# WRONG
lst = [1,2]
lst = lst.append(3)
## append returns None → lst becomes None

# CORRECT
lst = [1,2]
lst.append(3)
## lst is now [1,2,3]


# MISINTERPRETATION
df["A"] + df["B"]
## NOT column totals

# CORRECT INTERPRETATION
df["A"] + df["B"]
## element-wise (row-wise): result[i] = A[i] + B[i]


# WRONG
df.groupby("A").mean("B")
## incorrect syntax

# CORRECT
df.groupby("A")["B"].mean()
## select column first, then aggregate


########################
# LAST-SECOND DECISION GUIDE

## “Select rows…” → df[condition]
## “Create column…” → df["new"] = ...
## “Average by group…” → df.groupby("col")["target"].mean()
## “Count values…” → df["col"].value_counts()
## “Position-based selection…” → df.iloc[row, column]


########################
## pandas = row-wise operations
## iloc = [row, column]
## filtering = df[condition]
## groupby = split → apply → combine
```
