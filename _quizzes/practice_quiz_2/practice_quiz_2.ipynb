**Q2** Simulate a sequence of dice rolls

<font size = "3">

What you did for Q1 is essentially how the `random.choice` function works:


(You can also use the NumPy library and use `numpy.random.choice` instead).

Suppose we want to see how many rolls it takes to roll a "3" **twice** (in total, not consecutively). Write code that simulates this experiment. After your code completes, the variable `num_rolls` should be equal to the number of rolls it took to roll the second "3".

**Example:**

If we roll the following numbers (reading left-to-right):

$$ 1, 4, 3, 1, 1, 1, 2, 3$$

then `num_rolls` should have the value 8.

````python
num_rolls = 0


# SOLUTION
import random 
num_threes = 0 # how many threes we have rolled

while num_threes < 2:
    # roll the dice
    roll = random.choice([1, 2, 3, 4])

    # num_rolls should be increased by 1
    num_rolls = num_rolls + 1

    # did we roll a three?
    if roll == 3:
        # increase the number of threes we've seen by 1
        num_threes = num_threes + 1

# that's it, print the number of rolls
print("Number of rolls:", num_rolls)

````

**Q3** Simulate a sequence of dice rolls (this one's a bit harder)

<font size = "3">

Suppose we want to see how many rolls it takes to roll a "3" **two consecutive times**. Write code that simulates this experiment. After your code completes, the variable `num_rolls` should be equal to the number of rolls it took to roll two "3's" in a row.

**Example:**

If we roll the following numbers (reading left-to-right):

$$ 1, 4, 3, 1, 1, 1, 2, 3, 2, 4, 3, 3$$

then `num_rolls` should have the value 12.

**Hint:** The computer needs to "remember" the previous dice roll. The most straightforward approach is to define a variable `previous_3 = False` and switch it to `True` when a 3 is rolled (don't forget to switch it back if you don't get two in a row!)

````python
import random
previous_3 = False 
keep_going = True 

while keep_going:
    roll = random.choice([1, 2, 3, 4])
    num_rolls = num_rolls + 1

    if roll == 3:
        # did we roll a 3 previously?
        if previous_3:
            # we got two in a row, we can stop
            keep_going = False
        else:
            # this is the first 3, but let's "remember" we rolled it for next time
            previous_3 = True
    else:
        # Wasn't a 3, so let's "remember" we failed to roll a 3 for next time
        previous_3 = False

print("Number of rolls:", num_rolls)


````
