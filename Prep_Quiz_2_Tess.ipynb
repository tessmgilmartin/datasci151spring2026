```python
##SEQUENCE ROLLS, how many rolls to get 3 twice?
num_rolls = 0


# SOLUTION
import random 
num_threes = 0 # how many threes we have rolled

while num_threes < 2:
    # roll the dice
    roll = random.choice([1, 2, 3, 4])

    # num_rolls should be increased by 1
    num_rolls = num_rolls + 1

    # did we roll a three?
    if roll == 3:
        # increase the number of threes we've seen by 1
        num_threes = num_threes + 1

# that's it, print the number of rolls
print("Number of rolls:", num_rolls)
```

```python
##SEQUENCE OF DICE ROLLS, how many rolls to get three twice CONSECUTIVELY
num_rolls = 0

# SOLUTION
import random
previous_3 = False 
keep_going = True 

while keep_going:
    roll = random.choice([1, 2, 3, 4])
    num_rolls = num_rolls + 1

    if roll == 3:
        # did we roll a 3 previously?
        if previous_3:
            # we got two in a row, we can stop
            keep_going = False
        else:
            # this is the first 3, but let's "remember" we rolled it for next time
            previous_3 = True
    else:
        # Wasn't a 3, so let's "remember" we failed to roll a 3 for next time
        previous_3 = False

print("Number of rolls:", num_rolls)
```

```python
##USING BREAK STATEMENT IN WHILE 
# Specify a threshold level

eps = 0.025

x_list = []
y_list = []


x = 1

max_points = 100000

for k in range(max_points):
    if 1/x <= eps:
        break

    x_list.append(x)
    y_list.append(1/x)
    x = x + 1

plt.scatter(x= x_list, y = y_list)
plt.title("Graph of f(x) = 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
